```python

# List of DataFrames
dataframes = {
    "df_Hyp1_Gene": df_Hyp1_Gene,
    "df_Hyp1_Isoform": df_Hyp1_Isoform,
    "df_Hyp2LOE_Gene": df_Hyp2LOE_Gene,
    "df_Hyp2LOE_Isoform": df_Hyp2LOE_Isoform,
    "df_Hyp2GOE_Gene": df_Hyp2GOE_Gene,
    "df_Hyp2GOE_Isoform": df_Hyp2GOE_Isoform,
    "df_Hyp3GOE_Gene": df_Hyp3GOE_Gene,
    "df_Hyp3GOE_Isoform": df_Hyp3GOE_Isoform,
    "df_Hyp5_Gene": df_Hyp5_Gene,
}

percentiles = [99.9, 99.5, 99, 98, 95]

def calculate_ranks_for_sample(df):
    for sample in df['Sample'].unique():
        sample_df = df[df['Sample'] == sample].copy()

        # Loop through each percentile, calculate rank, and add to the DataFrame
        for percentile in percentiles:
            # Calculate the percentile value for z-scores within this sample
            # This will not work if there are any NaN z-scores.
            percentile_value = np.percentile(sample_df['z_score'], percentile)
            
            # Create a new column name for the rank at this percentile
            # Create a new column name for the rank at this percentile
            if percentile == 99.5:
                rank_column_name = 'rank_top_99_5_percentile'
            elif percentile == 99.9:
                rank_column_name = 'rank_top_99_9_percentile'
            else:
                rank_column_name = f'rank_top_{percentile}_percentile'
            
            # Initialize the column with NaN for the whole DataFrame
            if rank_column_name not in df.columns:
                df[rank_column_name] = np.nan
            
            # Filter the DataFrame to include only rows with z-scores above this percentile within the sample
            top_percentile_df = sample_df[sample_df['z_score'] > percentile_value]
            
            # Sort the filtered DataFrame by the test statistic in descending order and reset index for ranking
            sorted_df = top_percentile_df.sort_values(by='test_statistic', ascending=False).reset_index(drop=True)
            
            # Create a mapping of Isoform_PBid to rank
            rank_mapping = {isoform: rank + 1 for rank, isoform in enumerate(sorted_df['Isoform_PBid'])}

            # Update the ranks in the original DataFrame for the current sample
            df.loc[(df['Sample'] == sample) & (df['Isoform_PBid'].isin(rank_mapping.keys())), rank_column_name] = \
                df['Isoform_PBid'].map(rank_mapping)

    return df


# Loop through each DataFrame and apply the rank calculation
for name, df in dataframes.items():
    print(f"Calculating ranks for {name}...")
    dataframes[name] = calculate_ranks_for_sample(df)
    print(f"Finished calculating ranks for {name}.")

# Now, each DataFrame in `dataframes` has the new rank columns added.
```

```python
# Manually curated list of samples and their phenotypes (from seqr mostly)
probands = probands[['Sample','Phenotype']].drop_duplicates()
probands = probands[probands['Phenotype'].notnull()]

def stringToHPO(input_list,silent=True):
    """
    Matches a list of input term strings to HPO (Human Phenotype Ontology) terms.

    Parameters:
    input_list (list of str): List of input terms to be matched to HPO terms.
    silent (bool): If True, suppresses error messages when a term is not found. Default is True.
    as_string (bool): If you want to return the list of terms in string (human-readable) output rather than HPO. Default is False.
    Returns:
    list: List of matched HPO terms.
    """
    output_list = []
    for term in input_list:
        try:
            hpo_term = str(Ontology.get_hpo_object(term.strip())).split(" | ")[0]
            output_list.append(hpo_term)
        except RuntimeError:
            if not(silent):
                print(f"Not found {term}")
    return output_list
    


probands['hpo_terms'] = [stringToHPO(r['Phenotype'].strip().strip(",").split(",")) for i,r in probands.iterrows()] 
probands['hpo_set'] = [HPOSet.from_queries(x) for x in probands['hpo_terms']]

## The difference between terms and set is: 
## hpo_terms: This is a list of Human Phenotype Ontology (HPO) terms that have been matched from the phenotype descriptions. Each element in this list is a string representation of an HPO term. This representation is typically used for simple matching and display purposes.
## hpo_set: This is a more structured and comprehensive representation, which involves creating a set of HPO terms (HPOSet). This set can be used for more complex operations like computing similarities, aggregating information, or performing other ontology-based computations.
```

```python
# genes <-> diseases (omim)

## To generate genemap2_processed.tsv, a separate script needs to be run first: 
# /mmfs1/gscratch/stergachislab/emcarth/projects/undiag_multiome/data/OMIM/parseGenemap.py

## The end of the genemap2 file has this info on the phenotype code (we only want genes that have a code of 3):
# 1 - the disorder is placed on the map based on its association with
# a gene, but the underlying defect is not known.
# 2 - the disorder has been placed on the map by linkage; no mutation has
# been found.
# 3 - the molecular basis for the disorder is known; a mutation has been
# found in the gene.
# 4 - a contiguous gene deletion or duplication syndrome, multiple genes
# are deleted or duplicated causing the phenotype.

# This script processes the 'Phenotypes' column of the OMIM genemap2.txt file, extracting detailed information into new columns: 'PhenotypesExtracted_PhenotypeName', 'PhenotypesExtracted_OMIMnum', 'PhenotypesExtracted_code', and 'PhenotypesExtracted_inheritance'. It handles NaN values, splits the 'Phenotypes' column by semicolons, and uses regex to parse phenotype entries, creating new rows for multiple entries. 
# If a gene is associated with more than one phenotype, each phenotype get it's own new row in the processed file. 

# Function to extract phenotype details using regex
def extract_phenotype_details(phenotype):
    # Regex patterns
    phenotype_pattern = r'^(?:\{|\[)?(.*?)(?:\}|\])?, (\d{6}) \((\d+)\)(?:, )?(.*)?$'


    # Match the patterns
    match = re.match(phenotype_pattern, phenotype)
    

    if match:
        phenotype_name, omim_num, code, inheritance = match.groups()
    else:
        # If the pattern doesn't match, return NaNs
        phenotype_name, omim_num, code, inheritance = np.nan, np.nan, np.nan, np.nan

    if inheritance == "":
        inheritance = np.nan
    return phenotype_name, omim_num, code, inheritance

# List to hold the new rows
new_rows = []

# Process each row
for _, row in genemap.iterrows():
    phenotypes = row['Phenotypes']
    
    if pd.isna(phenotypes):
        new_row = row.to_dict()
        new_row.update({
            'PhenotypesExtracted_PhenotypeName': np.nan,
            'PhenotypesExtracted_OMIMnum': np.nan,
            'PhenotypesExtracted_code': np.nan,
            'PhenotypesExtracted_inheritance': np.nan
        })
        new_rows.append(new_row)
    else:
        for phenotype in phenotypes.split(';'):
            phenotype = phenotype.strip()
            phenotype_name, omim_num, code, inheritance = extract_phenotype_details(phenotype)
            new_row = row.to_dict()
            new_row.update({
                'PhenotypesExtracted_PhenotypeName': phenotype_name,
                'PhenotypesExtracted_OMIMnum': omim_num,
                'PhenotypesExtracted_code': code,
                'PhenotypesExtracted_inheritance': inheritance
            })
            new_rows.append(new_row)

# Create a new DataFrame with the new rows
new_df = pd.DataFrame(new_rows)

# Save the new DataFrame to a CSV file (if needed)
new_df.to_csv('genemap2_processed.tsv', sep="\t",index=False)

```

```python
# genes <-> diseases (omim) <-> phenotypes (hpo)
# Now add HPO terms to each of the omim diseases.

genemap = pd.read_csv(f"genemap2_processed.tsv", sep='\t')
genemap = genemap[genemap["PhenotypesExtracted_code"] == 3]
genemap = genemap[genemap['Approved Gene Symbol'].notnull()]
genemap = genemap[['Approved Gene Symbol','PhenotypesExtracted_OMIMnum']].drop_duplicates()

def readableHPOterm(input_list, silent=True):
    if input_list != input_list:
        return input_list
    else:
        output_list = []
        for term in input_list:
            try:
                hpo_term = str(Ontology.get_hpo_object(term.strip())).split(" | ")[1]
                output_list.append(hpo_term)
            except RuntimeError:
                if not(silent):
                    print(f"Not found {term}")
        
        return output_list
    
genemap['PhenotypesExtracted_OMIMnum'] = [f"OMIM:{int(x)}" if not(pd.isnull(x))  else x for x in genemap['PhenotypesExtracted_OMIMnum']]
genemap['HPO_terms'] = [hpo_dict[x] if (x in hpo_dict) else np.nan for x in genemap['PhenotypesExtracted_OMIMnum']]
genemap['HPO_terms_readable'] = [readableHPOterm(i) for i in genemap['HPO_terms']]
```

```python
## 1 = proband
## 2 = parent 1
## 3 = parent 2
def annotateInheritance(df):
    if df['num_alt_alleles_1'] == 0:
        return "not inherited"
    if (df['num_alt_alleles_1'] >= 1) and ((df['num_alt_alleles_2'] == 0) and (df['num_alt_alleles_3'] == 0)):
        return "de novo"
    if (df['num_alt_alleles_1'] == 2) and ((df['num_alt_alleles_2'] == 1) and (df['num_alt_alleles_3'] == 1)):
        return "recessive"
    if (df['num_alt_alleles_1'] == 1) and ((df['num_alt_alleles_2'] >= 1) or (df['num_alt_alleles_3'] >= 1)):
        return "from single parent"
    return "other"

data_types = ['all_nonsense', 'splicing','strict_missense','strict_regulatory']

seqr = {}

for i in data_types:
    df_UDN = pd.read_csv(f"{seqr_files_dir}/{i}_UDN.tsv",sep="\t")[['chrom','pos','ref','alt','gene','worst_consequence','exac_freq','cadd','splice_ai','family_id_1','num_alt_alleles_1','num_alt_alleles_2','num_alt_alleles_3','num_alt_alleles_4']]
    df_BCH = pd.read_csv(f"{seqr_files_dir}/{i}_BCH.tsv",sep="\t")[['chrom','pos','ref','alt','gene','worst_consequence','exac_freq','cadd','splice_ai','family_id_1','num_alt_alleles_1','num_alt_alleles_2','num_alt_alleles_3','num_alt_alleles_4']]
    df_BCH['family_id_1'] = [f"BCH_{x.split("_")[1]}-01" for x in df_BCH['family_id_1']] ## add -01 to the end of sample names.
    df = pd.concat([df_BCH, df_UDN],ignore_index=True)
    df['inheritance'] = [annotateInheritance(r) for i,r in df.iterrows()]
    ## df = df[(df['inheritance'] != 'from single parent') & (df['inheritance'] != 'not inherited')]
    df = df[(df['inheritance'] != 'not inherited')]
    #df.groupby('inheritance').count()['chrom']
    seqr[i] = df
    len(df)

```

```python
def process_isoform_data(isoform_df, probands, genemap, all_comparisons_long, combined_df, full_output_filename, filtered_output_filename):
    # Add proband HPO
    probands_selected = probands[['Sample', 'Phenotype', 'hpo_set']].rename(columns={'Phenotype': 'Proband_Phenotype', 'hpo_set': 'proband_hpo_set'})
    isoform_df = isoform_df.merge(probands_selected, how='left', left_on='Sample', right_on='Sample')
    
    # Add gene HPO
    isoform_df = isoform_df.merge(genemap, how='left', left_on='associated_gene', right_on='Approved Gene Symbol')
    
    # Add similarity score
    isoform_df = isoform_df.merge(all_comparisons_long, how='left', on=['PhenotypesExtracted_OMIMnum', 'Sample'])
    
    # Sort by similarity score
    isoform_df = isoform_df.sort_values('similarity', ascending=False)
    
    # If multiple phenotypes associated with the same gene, keep the one with the best phenotypic match
    isoform_df = isoform_df[~isoform_df[['associated_gene', 'Isoform_PBid', 'Sample']].duplicated()]
    
    # Create a set of (Sample, associated_gene) from combined_df
    combined_set = set(zip(combined_df['family_id_1'], combined_df['gene']))
    
    # Check if each row in isoform_df exists in the combined_set
    isoform_df['Associated_with_genetic_variant'] = isoform_df.apply(
        lambda row: (row['Sample'], row['associated_gene']) in combined_set, axis=1
    )
    
    # Fill NA/NaN values with False in the 'Associated_with_genetic_variant' column
    isoform_df['Associated_with_genetic_variant'] = isoform_df['Associated_with_genetic_variant'].fillna(False)
    
    columns_to_delete = [
        'Mouse Gene Symbol/ID', 'Approved Gene Symbol', 'HPO_terms', 'proband_hpo_set', 
        'Cyto Location', 'Computed Cyto Location', 'Gene/Locus And Other Related Symbols', 
        'Gene Name', 'Entrez Gene ID'
    ]
    
    # Check for the existence of each column and drop them if they exist
    columns_to_drop = [col for col in columns_to_delete if col in isoform_df.columns]
    if columns_to_drop:
        isoform_df = isoform_df.drop(columns=columns_to_drop, errors='ignore')
    
    # Rename specific columns
    isoform_df = isoform_df.rename(columns={'HPO_terms_readable': 'OMIM_HPO', 'Proband_Phenotype': 'Proband_HPO'})
    
    # Write results in csv
    isoform_df.to_csv(full_output_filename, index=False)
    filtered_isoform_df = isoform_df[isoform_df['Associated_with_genetic_variant']]
    filtered_isoform_df.to_csv(filtered_output_filename, index=False)

    return isoform_df, filtered_isoform_df
```

```python
# Creat a function to handle fusion genes. If the associated gene is GeneA_GeneB, then the HPO terms and Seqr variant matching won't work.

def handle_fusion_genes(df):
    """
    Handles fusion genes by splitting the 'associated_gene' column and creating new rows for each value separated by "_".
    
    Parameters:
    df (pandas.DataFrame): The input DataFrame containing 'Sample', 'associated_gene', and other columns.

    Returns:
    pandas.DataFrame: The DataFrame with exploded rows for each fusion gene.
    """
    # Add a new column 'original_associated_gene' which is a duplicate of 'associated_gene'
    df['original_associated_gene'] = df['associated_gene']
    
    # Ensure 'associated_gene' column is treated as strings, handling NaN values
    df['associated_gene'] = df['associated_gene'].astype(str)
    
    # Define a function to split genes and create new rows, retaining all other column values
    def split_genes(row):
        genes = row['associated_gene'].split('_')
        # Create a new row for each gene, keeping all other columns the same
        new_rows = [row.to_dict() for gene in genes]
        for new_row, gene in zip(new_rows, genes):
            new_row['associated_gene'] = gene
        return new_rows

    # Initialize an empty list to store the exploded rows
    exploded_rows = []
    
    i = 0
    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        # Print progress
        i += 1
        if i % 1000000 == 0:
            print(f"Processed {i} rows")
            
        if '_' in row['associated_gene']:
            # If 'associated_gene' contains '_', split the row and extend the exploded_rows list with the new rows
            exploded_rows.extend(split_genes(row))
        else:
            # If it does not, add the original row to the exploded_rows list
            exploded_rows.append(row.to_dict())
    
    # Create a new DataFrame from the exploded rows
    new_df = pd.DataFrame(exploded_rows)
    
    # Remove exploded_rows from memory to save space
    del exploded_rows
    
    return new_df
```
